```python
df = pd.read_csv('/Users/zzgr/Desktop/zzgrSandaeteuk/PROJECT GAS STATION/Gas_Station.csv')
df.drop(columns='Unnamed: 0', inplace=True)
df
```

```python
# 결측치 처리
temp = df[df['경유'] == '-']
temp
```

```python
dropit = temp[temp['휘발유'] == '-']
dropit
```

```python
dropit.reset_index(inplace=True)
```

```python
dropidxlist = dropit['index'].tolist()
```

```python
df.drop(index=dropidxlist, inplace=True)
```

```python
# column type 변환 (obj -> int)
df['고급휘발유'].replace('-','0', inplace=True)
df['휘발유'].replace('-','0', inplace=True)
df['경유'].replace('-','0', inplace=True)
df['실내등유'].replace('-','0', inplace=True)
```

```python
df['고급휘발유'] = df['고급휘발유'].astype(int)
df['휘발유'] = df['휘발유'].astype(int)
df['경유'] = df['경유'].astype(int)
df['실내등유'] = df['실내등유'].astype(int)
```
